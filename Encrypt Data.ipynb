```python
import pandas as pd
import numpy as np
import hashlib
```

```python
df = pd.read_csv('Requests-before-2023.csv')
```

```python
df.info()
```

```python
private_columns = ['Branch or Department', 'Requester', 'Technician']
```

```python
df[private_columns] = df[private_columns].astype(str)
```

```python
for name in private_columns:    
    df[name] = df[name].apply(
        lambda x: 
            hashlib.sha256(x.encode()).hexdigest()
    )
```

```python
df.head()
```

```python
df.to_csv('Requests-before-2023--ENCRYPTED.csv')
```
